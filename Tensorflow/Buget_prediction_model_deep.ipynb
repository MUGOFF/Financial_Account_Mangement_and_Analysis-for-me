```python
import pandas as pd
import numpy as np


dgb_out_df = pd.read_csv('preprocessed_data\DGB_preprecess_expense_05.CSV')
dgb_out_df.head()
```

```python
ibk_out_df = pd.read_csv('preprocessed_data\IBK_output_05.CSV')
ibk_out_df.head()
```

```python
input_all_df = pd.concat([dgb_out_df, ibk_out_df], axis=0, ignore_index=True)
input_all_df.head()
```

```python
input_all_df['datetime'] = pd.to_datetime(input_all_df['datetime'])
input_all_df.info()
```

```python
expense_monthly = input_all_df[input_all_df['special_budget'] == 0].groupby([input_all_df['datetime'].dt.year.rename('year'), input_all_df['datetime'].dt.month.rename('month')])['amount'].sum()
expense_monthly.head(5)
```

```python
expense_monthly_df = expense_monthly.reset_index().copy()
expense_monthly_df.head()
```

```python
import pandas as pd
import numpy as np


dgb_in_df = pd.read_csv('preprocessed_data\DGB_preprecess_inputs_03_drop_nonIncome.CSV')
dgb_in_df.info()
```

```python
ibk_in_df = pd.read_csv('preprocessed_data\IBK_income_02_dropSelf.CSV')
ibk_in_df.info()
```

```python
input_all_df_income = pd.concat([dgb_in_df, ibk_in_df], axis=0, ignore_index=True)
input_all_df_income['datetime'] = pd.to_datetime(input_all_df_income['datetime'])
display(input_all_df_income.info())
display(input_all_df_income.head())
```

```python
income_monthly = input_all_df_income.groupby([input_all_df_income['datetime'].dt.year.rename('year'), input_all_df_income['datetime'].dt.month.rename('month')])['delta_value'].sum()
income_monthly.head(5)
```

```python
income_monthly_df = income_monthly.reset_index().copy()
income_monthly_df.head()
```
