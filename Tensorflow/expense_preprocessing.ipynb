```python
import pandas as pd
import numpy as np
import chardet

file_path2016 = 'raw_data\IBK2016.CSV'
file_path2017 = 'raw_data\IBK2017.CSV'
rawdata = open(file_path2016, 'rb').read()
result = chardet.detect(rawdata)
encoding = result['encoding']
```

```python
ibk_2016 = pd.read_csv(file_path2016, encoding=encoding)
ibk_2017 = pd.read_csv(file_path2017, encoding=encoding)
```

```python
display(ibk_2016.head())
display(ibk_2017.head())
```

```python
display(pd.concat([ibk_2016, ibk_2017], axis=0))
```

```python
import glob

file_list = glob.glob('raw_data\IBK*.CSV')
dfs = [pd.read_csv(file, encoding=encoding) for file in file_list]

# 데이터프레임 연결
IBK_concat = pd.concat(dfs, axis=0, ignore_index=True)
IBK_concat.head()
```

```python
IBK_concat['datetime'] = pd.to_datetime(IBK_concat['거래일시'], format='%Y-%m-%d %H:%M:%S')
IBK_concat.head()
```

```python
IBK_concat = IBK_concat.sort_values('datetime',ascending=False)
IBK_concat['출금'].fillna('0', inplace=True)
IBK_concat['입금'].fillna('0', inplace=True)
IBK_concat.head()
```

```python
IBK_concat.to_csv('preprocessed_data\IBK_merged_00.CSV',index=False)
```
