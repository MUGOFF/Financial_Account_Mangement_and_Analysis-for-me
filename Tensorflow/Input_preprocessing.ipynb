### 기본 전처리

```python
import pandas as pd
import numpy as np
import chardet

file_path = 'raw_data\DGB_~20240405_INPUT.CSV'
rawdata = open(file_path, 'rb').read()
result = chardet.detect(rawdata)
encoding = result['encoding']
```

```python
RawData = pd.read_csv(file_path, encoding=encoding)
display(RawData)
```

```python
RawData_filled = RawData.copy()
RawData_filled['메모'].fillna(" ", inplace=True)
RawData_filled.head()
```

```python
RawData.head()
```

```python
RawData_filled['비고'].fillna(RawData_filled['거래종류'], inplace=True)
RawData_filled.head()
```

```python
RawData_Drop = RawData_filled.drop(columns=['NO','출금금액','거래점','거래후잔액'])
RawData_Drop.head()
```

```python
RawData_Drop['income'] = pd.to_numeric(RawData_Drop['입금금액'].str.replace(',' , ''))
RawData_Drop.info()
```

```python
RawData_Drop['datetime'] = pd.to_datetime(RawData_Drop['거래일시'], format='%Y-%m-%d [%H:%M:%S]')
RawData_Drop.info()
```

```python
RawData_Drop.head()
```

```python
RawData_Drop_second = RawData_Drop.drop(columns=['거래일시', '입금금액'])
RawData_Drop_second.rename(columns={'비고':'target', '거래종류' : 'type', '메모' : 'memo'}, inplace= True)
RawData_Drop_second.head()
```

```python
RawData_Drop_selfIncome = RawData_Drop_second[RawData_Drop_second['target'].str.contains('문의곤') == False]
RawData_Drop_selfIncome.head()
```

```python
RawData_Drop_second.to_csv('preprocessed_data\DGB_preprecess_inputs.CSV', index = False)
RawData_Drop_selfIncome.to_csv('preprocessed_data\DGB_preprecess_inputs_drop_self.CSV', index = False)
```
