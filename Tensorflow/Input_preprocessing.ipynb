## DGB

```python
import pandas as pd
import numpy as np

file_path = 'preprocessed_data\DGB_preprecess_income_00.CSV'
DGB_input = pd.read_csv(file_path)
DGB_input['datetime'] = pd.to_datetime(DGB_input['datetime'])
display(DGB_input.info())
```

```python
DGB_input['targetType'] = '수입'
DGB_input['account'].unique()
```

```python
DGB_self = DGB_input[
    (DGB_input['target'].str.contains('527118939271') == True) |
    (DGB_input['target'].str.contains('511228697205') == True) |
    (DGB_input['target'].str.contains('신한0883926') == True) |
    (DGB_input['target'].str.contains('국민0046417') == True) |
    ((DGB_input['target'] == '신한우윤희') & (DGB_input['datetime'].dt.month == 1) & (DGB_input['datetime'].dt.day == 13)) |
    (DGB_input['target'].str.contains('아빠돈') == True) |
    (DGB_input['target'].str.contains('아빠 돈') == True) |
    (DGB_input['target'].str.contains('농협휴가잘보내') == True) |
    (DGB_input['target'].str.contains('농협정현실') == True) |
    (DGB_input['target'].str.contains('문하은') == True) |
    (DGB_input['target'].str.contains('혜경이모') == True) |
    (DGB_input['target'].str.contains('할머니') == True) |
    (DGB_input['target'].str.contains('뮤직카우') == True) |
    (DGB_input['target'].str.contains('비전장학금') == True) |
    (DGB_input['target'].str.contains('문의곤') == True) |
    (DGB_input['target'].str.contains('나라사랑') == True) |
    (DGB_input['target'].str.contains('한투') == True)
    ]
DGB_input.loc[DGB_input['target'].isin(DGB_self['target']) == True, 'targetType'] = '이체'
len(DGB_self)
```

```python
DGB_returnExpense = DGB_input[
    (DGB_input['target'].str.contains('트래블월렛환불') == True) |
    (DGB_input['target'].str.contains('장주복') == True) |
    (DGB_input['target'].str.contains('서지범') == True) |
    (DGB_input['target'].str.contains('이홍민') == True) |
    (DGB_input['target'].str.contains('박도현') == True) |
    (DGB_input['target'].str.contains('김태완') == True) |
    (DGB_input['target'].str.contains('이재빈') == True) |
    (DGB_input['target'].str.contains('체크취소') == True) |
    (DGB_input['target'].str.contains('체크카드승인취소') == True) |
    (DGB_input['target'].str.contains('한국철도공사') == True) |
    (DGB_input['account'] == 'DGB_지급취소')
    ]
DGB_input.loc[DGB_input['target'].isin(DGB_returnExpense['target']) == True, 'targetType'] = '소비'
```

```python
DGB_dropSelf = DGB_input.loc[DGB_input['targetType'] == '수입']
DGB_moving = DGB_input.loc[DGB_input['targetType'] == '이체']
DGB_return = DGB_input.loc[DGB_input['targetType'] == '소비']
```

```python
DGB_dropSelf.to_csv('preprocessed_data\DGB_preprecess_income_01.CSV', index = False)
DGB_moving.to_csv('preprocessed_data\DGB_preprecess_intermove_input_01.CSV', index = False)
DGB_return.to_csv('preprocessed_data\DGB_preprecess_expense_return_01.CSV', index = False)
```

## IBK

```python
import pandas as pd
import numpy as np

IBK_income = pd.read_csv('preprocessed_data\IBK_income_01.CSV')
IBK_income['datetime'] = pd.to_datetime(IBK_income['datetime'])
IBK_income.info()
```

```python
IBK_income['거래구분'].unique()
```

```python
IBK_income[IBK_income['거래구분'] == "CC"]
```

```python
IBK_income_copy = IBK_income.copy()

IBK_income_copy['거래구분'] = IBK_income_copy['거래구분'].fillna(IBK_income['거래내용'])
IBK_income_copy['거래내용'] = IBK_income_copy['거래내용'].fillna(IBK_income['거래구분'])
IBK_income_copy['메모'] = IBK_income_copy['메모'].fillna(' ')

IBK_income = IBK_income_copy
```

```python
IBK_income['거래내용'].fillna('', inplace=True)
IBK_income['상대계좌예금주명'].fillna('', inplace=True)
IBK_income['거래구분'].fillna('', inplace=True)
IBK_income_dropSelf = IBK_income.loc[
    (IBK_income['거래내용'].str.contains('문의곤') == False) &
    (IBK_income['상대계좌예금주명'].str.contains('문의곤') == False)
    ]

IBK_income_dropSelf.head()
```

```python
IBK_income_dropSelf_copy = IBK_income_dropSelf.copy()

# 복사본을 수정
IBK_income_dropSelf_copy['account'] = IBK_income_dropSelf_copy['거래구분']
IBK_income_dropSelf_copy['target'] = IBK_income_dropSelf_copy['거래내용']
IBK_income_dropSelf_copy['memo'] = IBK_income_dropSelf_copy['메모']
IBK_income_dropSelf_copy['targetType'] = '수입'

# 원본 DataFrame에 복사본을 할당
IBK_income_dropSelf = IBK_income_dropSelf_copy
```

```python
IBK_income_dropSelf.info()
```

```python
IBK_income_dropCol = IBK_income_dropSelf.drop(columns=['No','거래일시','거래내용','상대계좌번호','출금','입금','상대은행','메모','거래구분','상대계좌예금주명','거래후 잔액','수표어음금액','송금메시지','CMS코드'])
IBK_income_dropCol = IBK_income_dropCol.sort_values('datetime',ascending=False)
IBK_income_dropCol.head()
```

```python
IBK_income_dropCol.to_csv('preprocessed_data\IBK_income_02_dropSelf.CSV', index= False)
```
